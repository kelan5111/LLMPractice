```python
import keras
from keras import layers
import numpy as np

# Training data - addition of numbers
train_data = np.array([
    [2, 2],
    [5, 2],
    [10, 9],
    [1, 2],
    [22, 55],
    [100, 3],
    [12, 3],
    [4, 4],
    [1999, 4],
    [10000, 1200],
    [12, 45],
    [65, 99],
    [12321, 123],
    [1331, 0],
    [343, 123],
    [109, 12],
    [1999, 199],
    [444, 444],
    [1, 1],
    [99, 99999]
])
train_label = np.array([4, 7, 19, 3, 77, 103, 15, 8, 2003, 11200, 57, 164, 12444, 1331, 466, 121, 2198, 888, 2, 100098])

# Testing data
test_data = np.array([
    [10, 10],
    [3, 1],
    [19, 2],
    [2, 9],
    [5, 5],
    [1000, 3]
])
test_label = np.array([20, 4, 21, 11])
```

```python
# Model
model = keras.Sequential(
    [
        layers.Dense(16),
        layers.Dense(16),
        layers.Dense(1)
    ]
)
# compiling model
model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)
```

```python
model.fit(train_data, train_label, epochs=100, batch_size=6)
```

```python
predictions = model.predict(test_data)
for i in range(len(test_data)):
    print(int(test_data[i][0:][0]), "+", int(test_data[i][0:][1]), "=", int(predictions[i][0:][0]))
```

```python
a = float(input("First number: "))
b = float(input("Second number: "))

pred = model.predict(np.array([[a, b]]), verbose=0)
print(f"{int(a)} + {int(b)} = {round(float(pred[0][0]))}")
```
