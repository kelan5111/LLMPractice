```python
# Tensor = container storing numerical data
# There are many different tensors:
# 1. scalar (rank-1)
# 2. vector (rank-2)
# 3. matrice (rank-3)
# 4. rank-n tensors

import numpy as np
```

```python
# Scalar tensor: single number (float32, float64)
x = np.array(12)
print(x)
print(x.ndim)
```

```python
# Vector tensor: 1D, one axis
x = np.array([12, 3, 6, 14, 7])
print(x)
print(x.ndim)
```

```python
# Matrices: 2D, two axis (rows & columns) - array of vectors
x = np.array([[1, 4, 10, 2],
             [4, 5, 1, 15],
             [10, 8, 2, 4]])
print(x)
print(x.ndim)
```

```python
# Rank-3 tensor
x = np.array([[[3, 2, 1, 4], [2, 4, 2, 1]],
              [[1, 4, 1, 3], [4, 3, 9, 2]]])
print(x)
print(x.ndim)
```

```python
# Tensor attribute:
# 1. axes (rank): ndim of a tensor - 2-rank has 2 axis
# 2. shape: dimension along each axis
# 3. data type: type of data stored in a tensor e.g. float-32, bool, uint8
```
