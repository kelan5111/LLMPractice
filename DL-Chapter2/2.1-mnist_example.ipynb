```python
from keras.datasets import mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# 4 different numpy arrays train_images, train_labels, test_images, test_labels- mnist is a training set of 60,000 training images plus 10,000 test images

# train_images, train_labels = training set - the data the model will learn from
# test_images, test_labels = test set - the data that the model will be tested on
```

```python
import keras
from keras import layers

# Building a network model
model = keras.Sequential(
    [
        layers.Dense(512, activation="relu"),   # Dense = fully connected layer
        layers.Dense(10, activation="softmax"), # softmax classification layer which returns a 10-way (summing to 1)
    ]
)

# Layer = building block of a neural network acts as a data filter
# - transforms input data through learned parameters (weights and biases)
```

```python
# Loss function = measures model performance
# Optimiser = mechanism used to update weights to improve model's performance

model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"],
              )
```

```python
# Preparing the image data
# Converting numpy arrays into the shape the model expects
train_images = train_images.reshape((60000, 28*28))
train_images = train_images.astype("float32") / 255
test_images = test_images.reshape((10000, 28*28))
test_images = test_images.astype("float32") / 255
```

```python
# Fitting the model to its training data
model.fit(train_images, train_labels, epochs=5, batch_size=128)

# accuracy = correct prediction
# loss = how wrong the model is in the moment
```

```python
# Using the model to make predictions using the test set

test_digits = test_images[0:10] # taking 10 test images from array
predictions = model.predict(test_digits)
predictions[0]
```

```python
# Evaluating the model on new data
test_loss, test_acc = model.evaluate(test_images, test_labels)
print(f"test acc: {test_acc}")
```

```python
# Conclusion:
# 1. Model makes predictions on a batch
# 2. Loss Function compares the predictions to the true labels -> produces the loss value
# 3. Optimizer (adam) uses that loss to nudge the weights via backpropagation
# 4. Repeat for every batch
```
